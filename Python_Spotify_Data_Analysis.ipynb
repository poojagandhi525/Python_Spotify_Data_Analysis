CORRELATION BETWEEN ENERGY AND LOUDNESS

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x = "energy", y = "loudness" , color = "c").set(title = "Correlation Between Energy and Loudness")
plt.show()
```

From the above graph, we can see that there is high correlation between energy and loudness.

Correlation Between Acousticness and Popularity

```python
plt.figure(figsize = (10,20))
sns.regplot(data = sample_df, x = "acousticness", y = "popularity" , color = "b").set(title = "Correlation Between Acousticness and Popularity")
plt.show()
```

From the above graph we can see that there is negative correlation in Acousticness and Popularity
